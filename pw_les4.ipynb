### Тема “Предел функции”

1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях. 
    
    $f(x)=x+\frac{1}{x}$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format='retina'

sns.set(rc={'figure.figsize': (9, 6)})
```

```python
x = np.arange(-10, 10, 0.125)
y = x +1/x

plt.axhline(color='r')
plt.axvline(color='r')

plt.plot(x,y)
plt.show()
```

### Тема “Теоремы о пределах”

1. Найти предел:

a. $\lim\limits_{x \to 0}\frac{sin\big(2x\big)}{4x} =\lim\limits_{x \to 0}\frac{sin\big(2x\big)}{2 \cdot 2x}=\lim\limits_{x \to 0}\frac{1}{2} \cdot \frac{sin\big(2x\big)}{2x} = \lim\limits_{x \to 0}\frac{1}{2} \cdot 1 = \frac{1}{2}$

b. $\lim\limits_{x \to 0}\frac{x}{sin\big(x\big)} = 1$

c. $\lim\limits_{x \to 0}\frac{x}{arcsin(x)} = 1$

d. $\lim\limits_{x \to \infty}\big(\frac{4x+3}{4x-3}\big)^{6x}=\lim\limits_{x \to \infty}\big(\frac{4x-3+6}{4x-3}\big)^{6x}=\lim\limits_{x \to \infty}\big(\frac{4x-3}{4x-3}+\frac{6}{4x-3}\big)^{6x}= \lim\limits_{x \to \infty}\big(1+ \frac{6}{4x-3}\big)^{6x}=\lim\limits_{x \to \infty}\big(1+ \frac{6}{4x-3}\big)^{\frac{4x-3}{6}\cdot\frac{6}{4x-3} \cdot 6x}= e^{\lim\limits_{x \to \infty}\big(\frac{6}{4x-3} \cdot 6x\big)}=e^{\lim\limits_{x \to \infty}\big(\frac{36x}{4x-3}\big)}= e^{\big(\frac{36}{4}\big)}= e^9$
